```python
import re

def preprocess(input_text):
    input_lower = input_text.lower()
    make_tokens = nltk.word_tokenize(input_lower)
    str_tokens = ' '.join(str(token) for token in make_tokens)
    stopwords_ = stopwords.words('english')
    tokens_without_stopwords = [token for token in make_tokens if token.lower() not in stopwords_]
    text_without_stopwords = ' '.join(tokens_without_stopwords)
    txt_no_punc = re.sub(r'[^\w\s]', '', text_without_stopwords)
    preprocess_text = txt_no_punc.replace("  ", " ")
    # print(preprocess_text)
    return preprocess_text
```

```python
def merge_lists(set1, set2, operation):
    comparison_made = 0
    set_result = set()

    if operation == 'AND':
        set_result = set1.intersection(set2)
        comparison_made += len(set1) + len(set2)  # Count comparisons as both sets are traversed

    elif operation == 'AND NOT':
        set_result = set1.difference(set2)
        comparison_made += len(set1) + len(set2)  # Count comparisons as both sets are traversed

    elif operation == 'OR':
        set_result = set1.union(set2)
        comparison_made += len(set1) + len(set2)  # Count comparisons as both sets are traversed

    elif operation == 'OR NOT':
        set_result = set1.union(set1, universal_set - set2)
        comparison_made += len(set1)  # Only set1 is traversed

    return set_result, comparison_made
```

```python
import re
import nltk

def process_query(query, operators, postings):
    total_comparisons = 0
    docs = []

    for i, op in enumerate(operators):
        op = op.strip()
        if i == 0:
            if query[i] in postings:
              word1_docs = postings[query[i]]
            else:
              word1_docs = []
            if query[i+1] in postings:
              word2_docs = postings[query[i+1]]
            else:
              word2_docs = []

            if op == "AND":
                docs, comparisons = merge_lists(set(word1_docs.copy()), set(word2_docs.copy()), 'AND')
            elif op == "OR":
                docs, comparisons = merge_lists(set(word1_docs.copy()), set(word2_docs.copy().copy()), 'OR')
            elif op == "AND NOT":
                docs, comparisons = merge_lists(set(word1_docs.copy()), set(word2_docs.copy().copy()), 'AND NOT')
            elif op == "OR NOT":
                docs, comparisons = merge_lists(set(word1_docs.copy()), set(word2_docs.copy().copy()), 'OR NOT')
            else:
                print("Invalid Query!!!")
                return None, None
        else:
            if query[i + 1] in postings:
                word2_docs = postings[query[i + 1]]
            else:
                word2_docs = []
            if op == "AND":
                docs, comparisons = merge_lists(set(docs.copy()), set(word2_docs.copy()), 'AND')
            elif op == "OR":
                docs, comparisons = merge_lists(set(docs.copy()), set(word2_docs.copy()), 'OR')
            elif op == "AND NOT":
                docs, comparisons = merge_lists(set(docs.copy()), set(word2_docs.copy()), 'AND NOT')
            elif op == "OR NOT":
                docs, comparisons = merge_lists(set(docs.copy()), set(word2_docs.copy()), 'OR NOT')
            else:
                print("Invalid Query!!!")
                return None, None
        total_comparisons += comparisons

    return docs, total_comparisons

N = int(input("Enter the number of queries: "))  # number of queries
queries = []
operator_list = []

for p in range(N):
    print("Query ", p + 1, ":")
    query = input("Enter input sentence: ")
    preprocessed_query = preprocess(query)  # Preprocess the query
    queries.append(nltk.word_tokenize(preprocessed_query))  # Tokenize the preprocessed query
    print(queries)
    operators = input("Enter operation sequence separated by commas: ").upper().split(", ")
    operator_list.append(operators)
    print(operator_list)

for k in range(N):
    query = queries[k]
    operators = operator_list[k]
    print("Query ", k + 1, ":", " ".join(f"{q} {op}" for q, op in zip(query[:-1], operators)), query[-1])
    docs, comparisons = process_query(query, operators, posting_lists)
    print(docs)
    if docs is not None:

        print("Number of documents retrieved for query ", k + 1, ": ", len(docs))
        print("Documents for query ", k + 1, ": ", docs)
    else:
        print("Query ", k + 1, "is invalid.")
```
